#### European Countries
Use set operations to determine which European countries are not in the European Union. Use the Country column.

- Checking the subset of countries in Europe


```python
# Check to confirm that the EU countries are a subset of countries in Europe
import pandas as pd

#Load Europe and EU through pandas
europe_df = pd.read_excel('Europe_and_EU.xlsx',sheet_name='Europe')
eu_df=pd.read_excel('Europe_and_EU.xlsx',sheet_name='EU')
```

```python
#cleaning country column in Europe_df
cleaned_europe=[]
for country in europe_df['Country']:
    country=country.strip()
    country=' '.join(country.split())
    country=country.title()
    cleaned_europe.append(country)
europe_df['Country']=cleaned_europe
```

```python
#cleaning country column in Eu_df
cleaned_eu=[]
for country in eu_df['Country']:
    country=country.strip() # removes leading spaces
    country=' '.join(country.split()) # removes extra spaces between words* how to keep words like uk
    country=country.title() # capitalizes each word
    cleaned_eu.append(country)
eu_df['Country']=cleaned_eu
```

```python
#Check to confirm that the EU countries are a subset of countries in Europe
#Convert to sets
europe_set=set(europe_df['Country'])
eu_set=set(eu_df['Country'])
```

```python
#check if eu is a subset of europe.
is_subset = eu_set.issubset(europe_set)
print("Are all EU countries in the Europe list?:", is_subset)
```

```python
#show which EU countries are missing from Europe list
missing = eu_set - europe_set
print("EU countries not found in Europe list:", missing) #display result
```

```python
#find countries in Europe but not in EU
europe_set = set(europe_df['Country']) #Convert to sets
eu_set = set(eu_df['Country'])
```

```python
#subtract EU from Europe
non_eu_countries = europe_set - eu_set
print("Countries in Europe but not in the EU:", non_eu_countries)# non EU countries.
```

```python
print("Number of countries in Europe but not in the EU:", len(non_eu_countries))
non_eu_countries = europe_set - eu_set
print(sorted(non_eu_countries))
```

```python
import pandas as pd
non_eu_df = pd.DataFrame(sorted(non_eu_countries), columns=['Country']) # convert set to a data frame
print(non_eu_df)
```
