```python
#C â (B â A) â C minus (B minus A)
#All elements in C that are not in B unless B is also in A
print("C â (B â A):", C - (B - A))
```

```python
#(C â© A) âª (C â B) â (C and A) or (C not B)
#All elements that are both in C and A, or in C but not in B
result = (C & A) | (C - B)
print("(C â© A) âª (C â B):", result)
```

#### Inclusion Exclusion Principle

```python
# Left-hand side: |A âª B âª C|
left_hand_eq = len(A | B | C)
print(left_hand_eq)  

# Right-hand side: |A| + |B| + |C| - |A â© B| - |A â© C| - |B â© C| + |A â© B â© C|
right_hand_eq = (
    len(A) + len(B) + len(C)
    - len(A & B) - len(A & C) - len(B & C)
    + len(A & B & C)
)
print(right_hand_eq)  

# Verify equality
print(left_hand_eq == right_hand_eq)  # Should print True
```

#### Writing down elements in a set

```python
#Create a set A of all possible outcomes assuming that all three cars drive in the same direction.
#All 3 must be the same:
A = {
    ("L", "L", "L"),
    ("S", "S", "S"),
    ("R", "R", "R")
}
```

```python
#Create a set B of all possible outcomes assuming that all three cars drive in a different direction
#All 3 are different
B = {
    ("L", "S", "R"),
    ("L", "R", "S"),
    ("S", "L", "R"),
    ("S", "R", "L"),
    ("R", "L", "S"),
    ("R", "S", "L")
}
```

```python
#Create a set C of all possible outcomes assuming that exactly 2 cars turn right
#This means 2 Râs and 1 different direction
C = {
    ("R", "R", "L"),
    ("R", "L", "R"),
    ("L", "R", "R"),
    ("R", "R", "S"),
    ("R", "S", "R"),
    ("S", "R", "R")
}
```

```python
#Create a set D of all possible outcomes assuming that exactly 2 cars drive in the same direction.
#This means one direction appears twice, and one other direction once - All same (A) All different (B)

D = {
    # Two Lefts:
    ("L", "L", "S"),
    ("L", "S", "L"),
    ("S", "L", "L"),
    
    # Two Straights:
    ("S", "S", "L"),
    ("S", "L", "S"),
    ("L", "S", "S"),
    
    # Two Rights: Already in C
    ("R", "R", "L"),
    ("R", "L", "R"),
    ("L", "R", "R"),
    ("R", "R", "S"),
    ("R", "S", "R"),
    ("S", "R", "R")
}
```

```python
# Write down the interpretation and give all possible outcomes for the sets denoted by:
#Dâ²
D_comp = A | B
D_comp
```
